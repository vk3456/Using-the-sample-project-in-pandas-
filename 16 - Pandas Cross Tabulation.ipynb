# Cross Tabulation
crosstab() function allows us to correlate two or more columns in our DataFrame.

```python
import pandas as pd

from pandas import DataFrame
```

```python
df = pd.read_csv('data_car_cross.csv',index_col=0)
df.head()
```

```python
pd.crosstab(df['cylinder'],df['year'])
```

```python
pd.crosstab(df['cylinder'],df['year'],margins=True)
```

```python
pd.crosstab(df['origin'],df['year'],margins=True)
```

```python
pd.crosstab(df['origin'],df['year'],rownames=['new_cylinder'],colnames=['new_year'])
```
